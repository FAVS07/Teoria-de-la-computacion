```python
import graphviz as gv
# Definir el autómata
automata = {
    'Star': {'': 'read',' ':'q0'},
    'read': {'': 'read', '': 'Star','':'q0'},
    'q0': {'1': 'q2', '0': 'q1'},
    'q1': {'1': 'q3', '0': 'q0'},
    'q2': {'1': 'q0', '0': 'q3'},
    'q3': {'1': 'q1', '0': 'q2'},
 
}

# Definir el estado inicial y los estados de aceptación
estado_inicial = 'Star'
estados_aceptacion = {'q0'}
```

```python
# Crear el objeto del grafo
g = gv.Digraph()

# Agregar los nodos
for estado in automata:
    if estado in estados_aceptacion:
        g.node(str(estado), shape='doublecircle')
    else:
        g.node(str(estado))

# Agregar las transiciones
for estado, transiciones in automata.items():
    for simbolo, destino in transiciones.items():
        g.edge(str(estado), str(destino), label=simbolo)


g.node('', shape='none')
g.edge('', str(estado_inicial), style='dashed')

# Mostrar el grafo
g.view()
```
