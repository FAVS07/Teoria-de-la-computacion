```python
import pandas as pd
import re
import numpy as np
```

```python
with open('numerosdeunos.txt', 'r') as file:
    contenido = file.read()
```

```python
filas = contenido.split("\n")
```

```python
ultimo_elemento=filas[len(filas)-1]
penultimo = filas[len(filas)-2]
ultimo_elemento = filas.pop()
penultimo = filas.pop()
```

```python
columnas = [fila.split(",") for fila in filas]
```

```python
import csv

with open('archivo.csv', mode='w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerows(columnas)
```

```python
mydata = pd.read_csv('archivo.csv')
```
